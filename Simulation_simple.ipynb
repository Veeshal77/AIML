```python
# Water runs through a faucet somwehere between 1 gallon per minute and 3 gallons per minute.
# How much time does it take to fill a 600 gallon pool?
```

```python
import random
import matplotlib.pyplot as plt

def fill_pool(size,Nexperiments=10000):
  flow_rate = []
  fill_time = []

  for i in range(Nexperiments):
    r = 1 + 2 * random.random()
    flow_rate.append(r)
    fill_time.append(size/r)

  return flow_rate, fill_time

size = 600
Nexperiments = 100000

flow_rate,fill_time = fill_pool(size,Nexperiments)

fig, (plt1,plt2) = plt.subplots(1,2,figsize=(10,5))
plt1.scatter(range(Nexperiments),flow_rate,s=1)
plt1.set_xlabel('Nexperiments')
plt1.set_ylabel('Flow rate')

plt2.scatter(range(Nexperiments),fill_time,s=1)
plt2.set_xlabel('Nexperiments')
plt2.set_ylabel('Fill time')

plt.show()

flow_rate.sort()
fill_time.sort()

fig2, (plt3,plt4) = plt.subplots(1,2,figsize=(10,5))
plt3.plot(flow_rate)
plt3.set_xlabel('Nexperiments')
plt3.set_ylabel('Flow rate')

plt4.plot(fill_time)
plt4.set_xlabel('Nexperiments')
plt4.set_ylabel('Fill time')

plt.show()


```
