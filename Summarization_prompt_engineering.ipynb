```python
# Using HuggingFace transformer classes
from transformers import AutoModelForSeq2SeqLM
from transformers import AutoTokenizer
from transformers import GenerationConfig

# load the FLAN-T5 model, creating an instance of the AutoModelForSeq2SeqLM class with the .from_pretrained() method

model_name = "google/flan-t5-large"
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

# download the tokenizer  from AutoTokenizer
tokenizer = AutoTokenizer.from_pretrained(model_name,use_fast=True)
```

```python
def get_completion(prompt):
  generation_config = GenerationConfig(max_new_tokens=100, do_sample=True, temperature=0.1)

  botInputs = tokenizer(prompt,return_tensors='pt').input_ids # pt => pytorch tensors

  botResponse = tokenizer.decode(
      model.generate(
          botInputs,
          generation_config = generation_config
      )[0],
      skip_special_tokens=True
  )

  return botResponse
```

```python
# Summary, Sentiment, Anger, Brand

for i in range(len(reviews)):
  prompt = f"""
  Generate a summary of the product review(s) in at the most 20 words.""" + reviews[i]
  response_Summary = get_completion(prompt=prompt)

  prompt = f"""
  Identify the sentiment of the reviews as positive, negative or neutral. Format your response as a JSON object with "Sentiment" as the key""" + reviews[i]
  response_sentiment = get_completion(prompt=prompt)

  prompt = f"""
  Is the reviewer expressing anger? Yes or No""" + reviews[i]
  response_anger = get_completion(prompt=prompt)

  prompt = f"""
  Identify the item being reviewed""" + reviews[i]
  response_item = get_completion(prompt=prompt)

  prompt = f"""
  Identify the brand from the review. If you cannot identify the brand just say NONE""" + reviews[i]
  response_brand = get_completion(prompt=prompt)

  print(f'Review{i} Summary: {response_Summary} \t Sentiment: {response_sentiment} \t Anger: {response_anger} \t Item: {response_item} Brand: {response_brand}\n')
```

```python
def get_completion_from_messages(prompt,temperature=0.1):
  generation_config = GenerationConfig(max_new_tokens=100, do_sample=True, temperature=temperature)

  botInputs = tokenizer(prompt,return_tensors='pt') # pt => pytorch tensors

  botResponse = tokenizer.decode(
      model.generate(
          botInputs["input_ids"],
          generation_config = generation_config
      )[0],
      skip_special_tokens=True
  )
  return botResponse
```
